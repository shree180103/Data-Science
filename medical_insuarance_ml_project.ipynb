Linear Regression with a single feature:

```python
from sklearn.linear_model import LinearRegression

model=LinearRegression()
```

```python
#now we are doing linear regression for non-smoker customers

non_smokers=medical_df[medical_df['smoker']=='no']

non_smokers
```

```python
inputs=non_smokers[['age']]
inputs
```

```python
targets=non_smokers[['charges']]


targets
```

```python
model.fit(inputs,targets)
```

```python
predictions=model.predict(inputs)
predictions
```

```python
targets
```

```python
from sklearn.metrics import mean_squared_error

rmse=mean_squared_error(targets,predictions,squared=False)

rmse
```

```python
w=model.coef_
w
```

```python
b=model.intercept_
b
```

```python
def LinearReg_plotter(I,P):
    plt.plot(I,P,'r')
    plt.scatter(I,targets,s=8)
    plt.xlabel('age')
    plt.ylabel('charges')


LinearReg_plotter(inputs,predictions)
```

now in the LinearRegression class, the optimizer used is ordinary least squares and not gradiesnt descent

now as we can need to standardize

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
numeric_cols=['age','bmi','children']
scaler.fit(medical_df[numeric_cols])
```

```python
scaled_input=scaler.transform(medical_df[numeric_cols])

scaled_input
```

```python
cat_cols=['smoker_numeric','sex_numeric','northeast','northwest','southeast','southwest']

data_categories=medical_df[cat_cols]

data_categories

```

```python
features=np.concatenate((scaled_input,data_categories),axis=1)
target_var=medical_df[['charges']]

model_scaled=LinearRegression().fit(features,target_var)
predictions_scaled=model_scaled.predict(features)

rmse=mean_squared_error(target_var,predictions_scaled,squared=False)

rmse

```

```python
model_scaled.coef_
```

```python
feature_cols=numeric_cols+cat_cols

  
d=pd.DataFrame(data={'features':np.append(feature_cols,'bias'),'weights':np.append(model_scaled.coef_,model_scaled.intercept_)})

d
```

now as we can see are weights make sense thus we can use then to explain our predictions

```python
features=np.concatenate((scaled_input,data_categories),axis=1)
target_var=medical_df[['charges']]

model=LinearRegression(positive=True).fit(features,target_var)
predictions_scaled=model.predict(features)

rmse=mean_squared_error(target_var,predictions_scaled,squared=False)

rmse
```

```python
from sklearn import linear_model
features=np.concatenate((scaled_input,data_categories),axis=1)
target_var=medical_df[['charges']]

model=linear_model.Ridge(alpha=0.00001).fit(features,target_var)
predictions_scaled=model.predict(features)

rmse=mean_squared_error(target_var,predictions_scaled,squared=False)

rmse
```

```python
from sklearn import linear_model
features=np.concatenate((scaled_input,data_categories),axis=1)
target_var=medical_df[['charges']]

model=linear_model.RidgeCV(alphas=np.logspace(-1, 6, 10)).fit(features,target_var)
predictions_scaled=model.predict(features)

rmse=mean_squared_error(target_var,predictions_scaled,squared=False)

rmse
```

```python
from sklearn import linear_model
features=np.concatenate((scaled_input,data_categories),axis=1)
target_var=medical_df[['charges']]

model=linear_model.Lasso(alpha=0.0001).fit(features,target_var)
predictions_scaled=model.predict(features)

rmse=mean_squared_error(target_var,predictions_scaled,squared=False)

rmse
```

```python
from sklearn.linear_model import LassoCV

features=np.concatenate((scaled_input,data_categories),axis=1)
target_var=medical_df[['charges']]

model=LassoCV(alphas=np.logspace(-10,10,13)).fit(features,target_var)
predictions_scaled=model.predict(features)

rmse=mean_squared_error(target_var,predictions_scaled,squared=False)

rmse
```

```python
from sklearn.linear_model import ElasticNet

features=np.concatenate((scaled_input,data_categories),axis=1)
target_var=medical_df[['charges']]

model=ElasticNet(alpha=0.001,l1_ratio=1).fit(features,target_var)
predictions_scaled=model.predict(features)

rmse=mean_squared_error(target_var,predictions_scaled,squared=False)

rmse
model.score(features,target_var)

from sklearn.metrics import r2_score
r2_score(target_var,predictions_scaled)
```

```python
from sklearn.linear_model import ElasticNetCV

features=np.concatenate((scaled_input,data_categories),axis=1)
target_var=medical_df[['charges']]

model=ElasticNetCV(cv=2).fit(features,target_var)
predictions_scaled=model.predict(features)

rmse=mean_squared_error(target_var,predictions_scaled,squared=False)

rmse
```

```python
baseline_y=[target_var.mean()]*len(target_var)
rmse=mean_squared_error(target_var,baseline_y,squared=False)

rmse
```

```python
from sklearn.metrics import r2_score

r2_score(target_var,baseline_y)
```

thus are model is better than a baseline model and has a score of 75%
